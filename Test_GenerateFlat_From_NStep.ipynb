# Generate MASK, FLAT, AND ZERO FRINGE for Ground Truth (FLAT)

```python
import scipy.io as sio
import numpy as np
from extras.FPP_functions import *
import cv2
import matplotlib.image as mpimg
from scipy.signal import hilbert
from unwrap import unwrap
from extras.ImgFunctions import *
import scipy.ndimage as ndi
import skimage.morphology as ski
from skimage.transform import pyramid_gaussian
import scipy.io as sio
from extras.dtcwt_denoiseFPP import dtcwt_denoiseFPP

# set 5
offsetrow  = 200
offsetcol = 500
sz = 1000


ftype = 'GT'
dir_ = '../CapImages/Set6/' + 'ref2'
N = 30
for i in range(N):
    if i == 0:       
        flat = loadAnImage( dir_ + '/' + '{0}_{1}.jpg'.format(ftype, i+1) )
        fringe = flat.copy()
    else:
        flat += loadAnImage( dir_ + '/' + '{0}_{1}.jpg'.format(ftype, i+1) )
flat = flat / N

imagesc(flat)
zeroFringe = fringe - flat


imagesc(zeroFringe, dpi_=300)
zeroFringe2 = dtcwt_denoiseFPP(zeroFringe, MaxIter=3)
imagesc(zeroFringe2, dpi_=300)


mask = np.abs(zeroFringe)>20
mask = ndi.binary_dilation(mask, iterations=10)
mask = ndi.binary_erosion(mask, iterations=10)
mask = ski.remove_small_objects(mask, min_size=20000)
        

imagesc(mask, dpi_=300)


rRoi, cRoi = mask.nonzero()
rowR = [np.min(rRoi), np.max(rRoi)]
colR = [np.min(cRoi), np.max(cRoi)]

# ## Save all to new files. 
# ftype = 'MASK'
# saveAnImage(dir_ + '/' + '{0}.jpg'.format(ftype), mask)

# ftype = 'FLAT'
# saveAnImage(dir_ + '/' + '{0}.jpg'.format(ftype), flat)

# ftype = 'GT_0'
# saveAnImage(dir_ + '/' + '{0}.jpg'.format(ftype), flat)


## load the FDB file and compare with the zeroFringe
ftype='FDB'
inpFringe = loadAnImage(dir_ + '/' + '{0}_{1}.jpg'.format(ftype, 1), isColor = 1)
plot(inpFringe[200:600, 1000,0], strFormat = 'r', dpi_=200)
plot(inpFringe[200:600, 1000,1], strFormat = 'g', dpi_=200)
plot(inpFringe[200:600, 1000,2], strFormat = 'b', dpi_=200)
plot(zeroFringe[200:600, 1000], strFormat = 'k', dpi_=200)
plot(zeroFringe2[200:600, 1000], strFormat = 'y', dpi_=200)
show()

ftype = 'GT'
res = {'GT_mask' : mask, 'GT_flat' : flat, 'GT_Fringe': zeroFringe2, 'GT_rRoi': rowR, 'GT_cRoi': colR, 'FDB_flat':inpFringe}
sio.savemat(dir_ + '/' + '{0}.mat'.format(ftype), res)
                
temp = sio.loadmat(dir_ + '/' + '{0}.mat'.format(ftype))
temp.keys()


```
